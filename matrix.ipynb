```python
import matplotlib.pyplot as plt
from PIL import Image
from numpy import asarray

# Load the first image and convert into a numpy array
img1 = Image.open('static/uploads/scaled_imageL.jpg')
numpydata1 = asarray(img1)

# Load the second image and convert into a numpy array
img2 = Image.open('static/uploads/scaled_image.jpg')
numpydata2 = asarray(img2)

# Ensure both arrays have the same shape
if numpydata1.shape == numpydata2.shape:
    # Calculate the pixel-wise absolute difference
    pixel_difference_absolute = abs(numpydata1 - numpydata2)

    # Display the original images
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(numpydata1, cmap='gray')
    plt.title('Lanczos')

    plt.subplot(1, 3, 2)
    plt.imshow(numpydata2, cmap='gray')
    plt.title('Billinear')

    # Display the pixel-wise absolute difference
    plt.subplot(1, 3, 3)
    plt.imshow(pixel_difference_absolute, cmap='viridis')  # Choose a colormap that suits your preference
    plt.title('Pixel-wise Absolute Difference')
    
    # Show the plot
    plt.show()
else:
    print("Image shapes are different. Cannot display pixel-wise difference.")
```

```python
import matplotlib.pyplot as plt
from scipy import ndimage
from skimage import data, color, io
import skimage.util.noise as noise
```

```python
image = io.imread("noisy.png")
gray_image = color.rgb2gray(image)

gn=noise.random_noise(image,mode='s&p')
result = ndimage.uniform_filter(gn, 7)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

ax = axes.ravel()
ax[0].imshow(gn, cmap='gray')
ax[0].set_title("Salt Pepper")
ax[1].imshow(result, cmap='gray')
ax[1].set_title("Low pass filter")
# ax[0].set_xlim(0, 512)
# ax[0].set_ylim(512, 0)
plt.tight_layout()
plt.show()
```

```python

import random 
import cv2 
  
def add_noise(img): 
  
    # Getting the dimensions of the image 
    row , col = img.shape 
      
    # Randomly pick some pixels in the 
    # image for coloring them white 
    # Pick a random number between 300 and 10000 
    number_of_pixels = random.randint(300, 10000) 
    for i in range(number_of_pixels): 
        
        # Pick a random y coordinate 
        y_coord=random.randint(0, row - 1) 
          
        # Pick a random x coordinate 
        x_coord=random.randint(0, col - 1) 
          
        # Color that pixel to white 
        img[y_coord][x_coord] = 255
          
    # Randomly pick some pixels in 
    # the image for coloring them black 
    # Pick a random number between 300 and 10000 
    number_of_pixels = random.randint(300 , 10000) 
    for i in range(number_of_pixels): 
        
        # Pick a random y coordinate 
        y_coord=random.randint(0, row - 1) 
          
        # Pick a random x coordinate 
        x_coord=random.randint(0, col - 1) 
          
        # Color that pixel to black 
        img[y_coord][x_coord] = 0
          
    return img 
  
# salt-and-pepper noise can 
# be applied only to grayscale images 
# Reading the color image in grayscale image 
img = cv2.imread('laut.webp', 
                 cv2.IMREAD_GRAYSCALE) 
  
#Storing the image 
cv2.imwrite('salt-and-pepper-lena.jpg', 
            add_noise(img))
```
